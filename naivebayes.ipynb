```python
import pandas as pd
```

```python
df = pd.read_csv('naivebayes2.csv')

df.head()
```

```python
def NaiveBayes(D:pd.DataFrame, **kwargs): # **kwargs: nhận vào các trường linh hoạt hơn
    # lấy ra nhãn ở cột cuối cùng
    last_feature = D.columns[-1]

    # Tính xác suất từng nhãn: P(Ci)
    labels = {}
    for label in D[last_feature]: labels[label] = 0
    for label in D[last_feature]: labels[label] += 1
    
    proportion = {}
    for key in labels.keys():
        proportion[key] = labels[key] / sum(labels.values())
    
    # Duyệt qua từng cột và tính xác suất từng điểm dữ liệu với nhãn đã xuất hiện
    res = {}
    for key in D[last_feature].unique(): # key: yes/no
        p = 1
        for f_name, f_value in kwargs.items():
            count_feature = D[(D[f_name] == f_value) & (D[last_feature] == key)].shape[0]
            p *= count_feature / labels[key]
        res[key] = p * proportion[key]
    result = max(res, key=res.get)
    
    kwargs[last_feature] = result
    
    D.loc[len(D)] = kwargs
    
    return result
```

```python
print(NaiveBayes(df, outlook='overcast', terrain='slope', temperature='mild', humidity='normal', wind='low'))
print(NaiveBayes(df, outlook='rainy', terrain='flat', temperature='cool', humidity='high', wind='weak'))
print(NaiveBayes(df, outlook='overcast', terrain='flat', temperature='mild', humidity='high', wind='low'))
print(NaiveBayes(df, outlook='overcast', terrain='undulating', temperature='cool', humidity='low', wind='strong'))
print(NaiveBayes(df, outlook='rainy', terrain='slope', temperature='cool', humidity='normal', wind='low'))
print(df)
```
